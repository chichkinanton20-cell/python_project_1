–í –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–æ–µ–∫—Ç–∞ –º—ã —Ä–µ–∞–ª–∏–∑—É–µ–º –∏–≥—Ä—É, –∫–æ—Ç–æ—Ä—É—é –≤—ã –±—É–¥–µ—Ç–µ —Å–æ–≤–µ—Ä—à–µ–Ω—Å—Ç–≤–æ–≤–∞—Ç—å –≤ —Ö–æ–¥–µ –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–æ–¥—É–ª—è. –í —ç—Ç–æ–º –±–ª–æ–∫–Ω–æ—Ç–µ –æ–ø–∏—Å–∞–Ω–æ –ø–æ—à–∞–≥–æ–≤–æ–µ –¢–ó, –≤—ã –º–æ–∂–µ—Ç–µ —Å–∫–∞—á–∞—Ç—å –µ–≥–æ –∫ —Å–µ–±–µ –≤ —Ä–µ–ø–æ–∑–∏—Ç–æ—Ä–∏–π –∏ –ø–∏—Å–∞—Ç—å –∫–æ–¥ –ø—Ä—è–º–æ –∑–¥–µ—Å—å, –∏–ª–∏ –∂–µ —Å–æ–∑–¥–∞—Ç—å –Ω–æ–≤—ã–π —Ñ–∞–π–ª —É —Å–µ–±—è. –ü–æ—Å–ª–µ —Ç–æ–≥–æ –∫–∞–∫ –∑–∞–ø—É—à–∏—Ç–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–Ω–æ–µ –∏–ª–∏ —á–∞—Å—Ç–∏—á–Ω–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ, –ø–∏—à–∏—Ç–µ –º–Ω–µ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏. –õ—é–±—ã–µ –≤–æ–ø—Ä–æ—Å—ã –º–æ–∂–µ—Ç–µ —Ç–∞–∫–∂–µ –∑–∞–¥–∞–≤–∞—Ç—å –≤ –ª–∏—á–∫–µ –∏–ª–∏ –ø—Ä—è–º–æ –≤ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è—Ö –≤ –∫–æ–¥–µ.

–î–ª—è –∏–≥—Ä—ã –≤—ã –º–æ–∂–µ—Ç–µ –ø—Ä–æ–¥—É–º–∞—Ç—å –ø–æ–ª–Ω–æ—Å—Ç—å—é —Å–≤–æ–π —Ñ–æ—Ä–º–∞—Ç, –∑–¥–µ—Å—å —è –Ω–∞–∫–∏–¥–∞—é —Å–∫–µ–ª–µ—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞. –†–µ–±—è—Ç–∞ —Ä–µ–∞–ª–∏–∑—É—é—Ç —Å–≤–æ–∏ –≤–∞—Ä–∏–∞–Ω—Ç—ã, –Ω–∞–ø—Ä–∏–º–µ—Ä –∞–≤—Ç–æ–±–∞—Ç–ª–µ—Ä –≥–¥–µ –±–∏—Ç–≤–∞ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø–∞—Ç–∏ –Ω–∞ –ø–∞—Ç–∏, –∫–∞–∫ –≤ autochess, super auto pets, hearthstone: battlegrounds, –∏ —Ç–¥. –ï—Å–ª–∏ —Å–æ–±–∏—Ä–∞–µ—Ç–µ—Å—å –¥–µ–ª–∞—Ç—å —Å–≤–æ—é –∏–¥–µ—é, –æ–ø–∏—à–∏—Ç–µ –µ—ë –≤ —Å–ª–µ–¥—É—é—â–µ–π —è—á–µ–π–∫–µ –≤–º–µ—Å—Ç–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –¢–ó.

### 1 –û–ø—Ä–µ–¥–µ–ª—è–µ–º –ø–µ—Ä—Å–æ–Ω–∞–∂–∞ –∏ –µ–≥–æ –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–æ–≤  

```python
hero_stats = {
    "health": 100,
    "attak": 25,
    "defense": 50,
    "X_place": 0,
    "Y_place": 0,
    "hero_symbol": "üßô"
}

enemies_coordinates = [
    [1,2],
    [2,2],
    [4,3]
]

enemy_hp = 80
enemy_atk = 15
enemy_def = 25
enemy_symbol = 'üëπ'

heal_coordinates = [
    [4,6],
    [1,7]
]

heal_symbol = "üçé"
heal_values = 10
```

### 2 –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ä–∞–∑–º–µ—Ä—ã –∫–∞—Ä—Ç—ã

```python
map_height=5
map_width=15
map_symbol='üå≥'
```

### 3 –§—É–Ω–∫—Ü–∏—è –æ—Ç—Ä–∏—Å–æ–≤—ã–≤–∞—é—â–∞—è –ø–æ–ª—è

```python
def draw(character_X, character_Y,  enemies_coordinates):
    
    print("–ö–∞—Ä—Ç–∞")
    
    map = []

    for i in range(map_height):
        map_st = []
        for j in range(map_width):
            map_st.append(map_symbol)
        map.append(map_st)

    map[character_Y][character_X] = hero_stats["hero_symbol"] #–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≥–µ—Ä–æ—è  
    
    for en in enemies_coordinates:
        map[en[0]][en[1]] = enemy_symbol #–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–æ–≤ 

    for heal_iter in heal_coordinates:
        map[heal_iter[0]][heal_iter[1]] = heal_symbol #–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∞–ø—Ç–µ—á–µ–∫ 

    for map_iter in map:
        print("".join(map_iter))
    
    
```

### 4   –§—É–Ω–∫—Ü–∏—è –ø–æ–¥—Å—á—ë—Ç–∞ —É—Ä–æ–Ω–∞

```python
def calculate_damage(attack, target_defence):
    if attack - (attack/100)*target_defence >= 0: #–¥–æ–±–∞–≤–∏–ª —á—Ç–æ–±—ã –Ω–µ –±—ã–ª–æ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–≥–æ —É—Ä–æ–Ω–∞, –µ—Å–ª–∏ –±—Ä–æ–Ω—è —Å–∏–ª—å–Ω–æ –±–æ–ª—å—à–µ —á–µ–º –∞—Ç–∞–∫–∞ 
        return attack - (attack/100)*target_defence
    else:
       return 0 
```

### 5 –§—É–Ω–∫—Ü–∏—è –±–∏—Ç–≤—ã 

```python
def fight(character_hp:int,
          character_atk:int,
          character_def:int,
          enemy_hp_fight:int,
          enemy_atk_fight:int,
          enemy_def_fight:int,
          verbose:bool):
    
    if verbose:print("–ù–∞—á–∏–Ω–∞–µ—Ç—Å—è –±–æ–π!")

    while True:
        
       #—É–¥–∞—Ä –ø–µ—Ä—Å–æ–Ω–∞–∂–∞

       enemy_hp_fight -= calculate_damage(character_atk, enemy_def_fight)
       if verbose: print(f"–ü–µ—Ä—Å–æ–Ω–∞–∂ –±—å—ë—Ç —Å –∞—Ç–∞–∫–æ–π {character_atk} –ø—Ä–æ—Ç–∏–≤ –∑–∞—â–∏—Ç—ã {enemy_def_fight}, –Ω–∞–Ω–æ—Å–∏—Ç {calculate_damage(character_atk, enemy_def_fight)} —É—Ä–æ–Ω–∞")
       if enemy_hp_fight == 0 or enemy_hp_fight < 0:
           if verbose: print(f"–ü–æ–±–µ–¥–∞! –£ –ø–µ—Ä—Å–æ–Ω–∞–∂–∞ –æ—Å—Ç–∞–ª–æ—Å—å {character_hp}")
           return character_hp
           break
       
       #—É–¥–∞—Ä –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞

       character_hp -= calculate_damage(enemy_atk_fight, character_def)
       if verbose: print(f"–ü—Ä–æ—Ç–∏–≤–Ω–∏–∫ –±—å—ë—Ç —Å –∞—Ç–∞–∫–æ–π {enemy_atk_fight} –ø—Ä–æ—Ç–∏–≤ –∑–∞—â–∏—Ç—ã {character_def}, –Ω–∞–Ω–æ—Å–∏—Ç {calculate_damage(enemy_atk_fight, character_def)} —É—Ä–æ–Ω–∞")
       if character_hp == 0 or character_hp < 0:
           if verbose: print("–ü–æ—Ä–∞–∂–µ–Ω–∏–µ! –ò–≥—Ä–∞ –æ–∫–æ–Ω—á–µ–Ω–∞!")
           return 0
           break
       
    if verbose:print(f"–£ –≥–µ—Ä–æ—è –æ—Å—Ç–∞–ª–æ—Å—å {character_hp} –æ—á–∫–æ–≤ –∑–¥–æ—Ä–æ–≤—å—è")
       
  
```

### 6 –¶–∏–∫–ª –ø–µ—Ä–µ–º–µ—â–µ–Ω–∏—è 

```python

control = ["w", "s", "a", "d", "q", "W", "S", "D", "A", "Q"]

draw(hero_stats["X_place"], hero_stats["Y_place"], enemies_coordinates=enemies_coordinates)

user_input = ''


while user_input != 'q':
    

    user_input = input()

     #—É–ø—Ä–∞–≤–ª–µ–Ω–µ–∏ —Å –ø–æ–º–æ—â—å—é wasd –∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è, —á—Ç–æ–±—ã –Ω–µ –≤—ã—Ö–æ–¥–∏–ª –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –∫–∞—Ä—Ç—ã 
    if user_input == "w" and hero_stats["Y_place"] != 0:
        hero_stats["Y_place"] -= 1
    elif user_input == "s" and hero_stats["Y_place"] != map_height - 1:
         hero_stats["Y_place"] += 1
    elif user_input == "a" and hero_stats["X_place"] != 0:
         hero_stats["X_place"] -= 1
    elif user_input == "d" and hero_stats["X_place"] != map_width - 1:
         hero_stats["X_place"] +=1
    elif user_input not in control:
         print("–¢–∞–∫–æ–π –∫–æ–º–∞–Ω–¥—ã –Ω–µ—Ç, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –¥—Ä—É–≥—É—é!","(w-–≤–ø–µ—Ä—ë–¥, s-–Ω–∞–∑–∞–¥, a-–≤–ª–µ–≤–æ, d-–≤–ø—Ä–∞–≤–æ)", sep = "\n")
         continue
    
    #–ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å—Ç–æ–ª–∫–Ω–æ–≤–µ–Ω–∏–µ —Å –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–æ–º
    for conf in enemies_coordinates:
          if hero_stats["X_place"] == conf[1] and hero_stats["Y_place"] == conf[0]:
               print("–°—Ç–æ–ª–∫–Ω–æ–≤–µ–Ω–∏–µ —Å –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–æ–º!")
               hero_stats["health"] = fight(character_hp = hero_stats["health"],
                    character_atk = hero_stats["attak"],
                    character_def = hero_stats["defense"],
                    enemy_hp_fight = enemy_hp,
                    enemy_atk_fight = enemy_atk,
                    enemy_def_fight = enemy_def,
                    verbose = True)
               if hero_stats["health"] != 0:
                    del enemies_coordinates[enemies_coordinates.index(conf)] #—É–±–∏–≤–∞–µ–º –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞(—É–±–∏—Ä–∞–µ–º —Å –∫–∞—Ä—Ç—ã)

    if hero_stats["health"] == 0 or len(enemies_coordinates) == 0:
           print("–ö–æ–Ω–µ—Ü –ò–≥—Ä—ã")
           break

    #–ª–µ—á–µ–Ω–∏–µ 
    for med in heal_coordinates:           
     if hero_stats["X_place"] == med[1] and hero_stats["Y_place"] == med[0] and hero_stats["health"] < 100:
              hero_stats["health"] += heal_values
              del heal_coordinates[heal_coordinates.index(med)]
              print(f"–ì–µ—Ä–æ–π –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞–ª—Å—è –∞–ø—Ç–µ—á–∫–æ–π! –û—Å—Ç–∞–ª–æ—Å—å {hero_stats["health"]} –æ—á–∫–æ–≤ –∑–¥–æ—Ä–æ–≤—å—è.")

              
                     
     
         
         


    draw(hero_stats["X_place"], hero_stats["Y_place"], enemies_coordinates) #–æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –ø–æ–ª—è
    

   
```
