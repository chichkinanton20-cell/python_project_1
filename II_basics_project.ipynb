```python
from IPython.display import clear_output
from config import MAP_HEIGHT, MAP_WIDTH, HEAL_SYMBOL
from scripts import Hero, World


world = World()

control = ["w", "s", "a", "d", "q", "e", "r"]

world.draw(world.X_place, world.Y_place)

user_input = ''


while user_input != 'q':
    
    clear_output(wait=True)#–æ—á–∏—â–∞–µ–º –∫–æ–Ω—Å–æ–ª—å
    user_input = input()

     #—É–ø—Ä–∞–≤–ª–µ–Ω–µ–∏ —Å –ø–æ–º–æ—â—å—é wasd –∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è, —á—Ç–æ–±—ã –Ω–µ –≤—ã—Ö–æ–¥–∏–ª –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –∫–∞—Ä—Ç—ã 
    if user_input.lower() == "w" and world.Y_place != 0:
        world.Y_place -= 1
    elif user_input.lower() == "s" and world.Y_place != MAP_HEIGHT - 1:
         world.Y_place += 1
    elif user_input.lower() == "a" and world.X_place != 0:
         world.X_place -= 1
    elif user_input.lower() == "d" and world.X_place != MAP_WIDTH - 1:
         world.X_place +=1
    elif user_input.lower() == "e":
         world.inventory_contents()
    elif user_input.lower() == "r":
         world.healing()
    elif user_input.lower() not in control:
         world.draw(world.X_place, world.Y_place,)
         print("–¢–∞–∫–æ–π –∫–æ–º–∞–Ω–¥—ã –Ω–µ—Ç, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –¥—Ä—É–≥—É—é!","(w-–≤–ø–µ—Ä—ë–¥, s-–Ω–∞–∑–∞–¥, a-–≤–ª–µ–≤–æ, d-–≤–ø—Ä–∞–≤–æ, e-–ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –∏–Ω–≤–µ–Ω—Ç–∞—Ä—å, r-–∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∞–ø—Ç–µ—á–∫—É)", sep = "\n")
         continue
    
    #–ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å—Ç–æ–ª–∫–Ω–æ–≤–µ–Ω–∏–µ —Å –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–æ–º
    world.conflict()

    if world.health == 0 or len(world.dragon_coordinats) == 0:
           print("–ö–æ–Ω–µ—Ü –ò–≥—Ä—ã")
           break

    #–≤–∑—è—Ç—å –∞–ø—Ç–µ—á–∫—É  
    for med in world.heal_coordinates:           
     if world.X_place == med[1] and world.Y_place == med[0]:
              del world.heal_coordinates[world.heal_coordinates.index(med)]
              world.inventory[HEAL_SYMBOL] += 1
              print(f"–ì–µ—Ä–æ–π –ø–æ–¥–æ–±—Ä–∞–ª –∞–ø—Ç–µ—á–∫—É!")

     #–≤–∑—è—Ç—å –∞—Ä—Ç–µ—Ñ–∞–∫—Ç—ã
     world.artifact_selection()

     world.city()         
                     
     
         
         


    world.draw(world.X_place, world.Y_place) #–æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –ø–æ–ª—è
    

   
```
