# TODO создайте множество из списка words

```python
words = ["apple", "plea", "leap", "pale", "peal", "bat", "tab", "cat", "act", "kek"]
given_word = set("bat") # выбранное слово
some_set = set(words)
```

## TODO создайте множество из слов у которых совпадает хотя бы одна буква с выбранным словом

```python
coincidence_set = set([i for i in words if set(i) & given_word])
print(coincidence_set)
```

# TODO множество слов, все буквы в которых присутствуют в выбранном слове

```python
coincidence_set = set([i for i in words if set(i) <= given_word])
print(coincidence_set)
```

# TODO множество слов в котором присутствуют все буквы из выбранного слова


```python
coincidence_set = set([i for i in words if set(i) >= given_word])
print(coincidence_set)
```
